```python
import pandas as pd

datapath1 = "Resources/schools_complete.csv"
datapath2 = "Resources/students_complete.csv"

school_data = pd.read_csv(datapath1)
student_data = pd.read_csv(datapath2)
school_data.head()
```

```python
student_data.head()
```

```python
# data_file1["school_name"].value_counts(), to check if there is any repeat value
total_school = len(school_data["school_name"])
total_student = len(student_data["Student ID"])
# I did not use the student_name because there might be students share the same name,
# However, the student ID should be unique
total_budget = school_data["budget"].sum()
ave_math_score = student_data["math_score"].mean()
ave_reading_score = student_data["reading_score"].mean()
```

```python
a = 0
for x in student_data["math_score"]:
    if x >= 70:
    # >=60 means pass or >=70 means pass?
        a += 1
math_passing_rate = a/len(student_data["math_score"])
```

```python
b = 0
for x in student_data["reading_score"]:
    if x >= 70:
        b += 1
reading_passing_rate = b/len(student_data["reading_score"])
```

```python
overall_passing_rate = (math_passing_rate + reading_passing_rate)/2
```

```python
district_summary = [total_school]
district_summary = pd.DataFrame(district_summary, columns={"Total Schools"})
district_summary["Total Students"] = total_student
district_summary["Total Budget"] = total_budget
district_summary["Average Math Score"] = ave_math_score
district_summary["Average Reading Score"] = ave_reading_score
district_summary["% Passing Math"] = math_passing_rate*100
district_summary["% Passing Reading"] = reading_passing_rate*100
district_summary["Overall Passing Rate"] = overall_passing_rate*100
district_summary
```

```python
new0 = school_data.set_index("school_name")
new1 = new0.drop(["School ID", "size", "budget"], 1)
new1.head()
```

```python
student_count = student_data.groupby("school_name").count()
# school_budget = student_data.groupby("budget").count()
```

```python
new1["Total Student"] = student_count["Student ID"]
new1["Total School Budget"] = new0["budget"]
new1["Per Student Budget"] = new0["budget"]/student_count["Student ID"]
```

```python
# new1["Total School Budget"] = new1["Total School Budget"].map("{:,}".format)
new1["Total School Budget"] = new1["Total School Budget"].map("${:.2f}".format)
## how to change the format for the second time?
new1["Per Student Budget"] = new1["Per Student Budget"].map("${:.2f}".format)
new1.head()
```

```python
new2 = student_data.set_index("school_name")
new2.head()
```

```python
new3 = new1.copy()
for i in list(school_data["school_name"]):
    score_list = new2.loc[i, ["math_score","reading_score"]]
    
    ave_math_score = score_list["math_score"].mean()
    ave_reading_score = score_list["reading_score"].mean()

    math_score_pass = score_list.loc[score_list["math_score"] >= 70, :]
    reading_score_pass = score_list.loc[score_list["reading_score"] >= 70, :]
    math_pass_rate = len(math_score_pass)/len(score_list)*100
    reading_pass_rate = len(reading_score_pass)/len(score_list)*100
    
    new3.loc[i, "Average Math Score"] = ave_math_score
    new3.loc[i, "Average Reading Score"] = ave_reading_score
    new3.loc[i, "% Passing Math"] = math_pass_rate
    new3.loc[i, "% Passing Reading"] = reading_pass_rate
    new3.loc[i, "% Overall Passing Rate"] = (math_pass_rate + reading_pass_rate)/2
    
new4 = new3.sort_values(by="% Overall Passing Rate", ascending=False)
new4.head(5)
```
