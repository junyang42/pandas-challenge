```python
import pandas as pd

datapath = "Resources/purchase_data.csv"
data_file = pd.read_csv(datapath)
data_file.head(5)
# data_file.count()
```

```python
# Q1

# data_file.shape
# list1 = data_file["SN"].value_counts()
# total1 = len(list1)

list_playername = data_file["SN"].unique()
total_player = len(list_playername)

#both total1 and total2 have the same answer

new1 = pd.DataFrame({"Total Players":[total_player]})
new1
```

```python
# Q2

unique_item_list = data_file["Item ID"].value_counts()
total_items = len(unique_item_list)
# total_items
```

```python
average_purchase_price = data_file["Price"].mean()
average_purchase_price = "$" + str(round(average_purchase_price, 2))
# average_purchase_price
```

```python
total_number_purchas = len(data_file["Price"])
# total_number_purchas
```

```python
total_revenue = data_file["Price"].sum()
total_revenue = "$" + str(round(total_revenue, 2))
# total_revenue
```

```python
new2 = pd.DataFrame({"Number of Unique Items": total_items,
                    "Average Purchase Price":average_purchase_price,
                    "Total Number of Purchases":total_number_purchas,
                    "Total Revenue":[total_revenue]})
# new2.dtypes
new2
```

```python
# Q3

player_gender_list = data_file["Gender"]
# player_gender_list.head()
male_player = data_file.loc[data_file["Gender"] == "Male", :]
male_player_list = male_player["SN"].unique()
male_player_number = len(male_player_list)
male_player_number
```

```python
female_player = data_file.loc[data_file["Gender"] == "Female", :]
female_player_list = female_player["SN"].unique()
female_player_number = len(female_player_list)
female_player_number
```

```python
Other_NonDisclosed_player = (total_player - male_player_number - female_player_number)
Other_NonDisclosed_player
```

```python
new3 = pd.DataFrame({"Total Count":[male_player_number, female_player_number, Other_NonDisclosed_player],
                     "Percentage of Players":[male_player_number/total_player, female_player_number/total_player, Other_NonDisclosed_player/total_player]})
new3.index = ["Male", "Female", "Other / Non-Disclosed"]
new3["Percentage of Players"] = (new3["Percentage of Players"]*100).round(2).astype(str) + "%"
new3
```

```python
# Q4

new4 = data_file[["Gender","Price"]]
new4 = new4.groupby("Gender")

purchase_count = new4.count()

average_purchase_p = new4.mean()
average_purchase_p = "$" + average_purchase_p.round(2).astype(str) 
total_purchase_v = new4.sum()
total_purchase_value = "$" + total_purchase_v.round(2).astype(str) 
```

```python
new_df = purchase_count.copy()
for i in list(purchase_count.index.values):
    
    new_df.loc[i,"Average Purchase Price"] = average_purchase_p.loc[i, "Price"]
    new_df.loc[i,"Total Purchase Value"] = total_purchase_value.loc[i, "Price"]
    
    ave_purchase_pp = "$" + (total_purchase_v.loc[i, "Price"]/new3.loc[i, "Total Count"]).round(2).astype(str)
    # (total_purchase_v.loc[i, "Price"]/new3.loc[i, "Total Count"]) gives us the average total purchase per person,
    new_df.loc[i,"Avg Total Purchase per Person"] = ave_purchase_pp
    
new_df.head()
```

```python
# Q5

bins = [-1,9,14,19,24,29,34,39, (data_file["Age"].max() + 1)]
group_list = ["<10", "10-14", "15-19", "20-24","25-29", "30-34", "35-39", "40+"]

new5 = data_file.copy()
new5["Age Range"] = pd.cut(new5["Age"], bins, labels=group_list)

unique_name_count = new5.groupby("Age Range")["SN"].nunique()
unique_name_count = pd.DataFrame(unique_name_count)
unique_name_count = unique_name_count.rename(columns={"SN":"Total Count"})
unique_name_count
```

```python
new5 = new5[["Age Range","Price"]]
new5 = new5.groupby("Age Range")
# new5.head()
purchase_count2 = new5.count()
average_purchase_p2 = new5.mean()
total_purchase_v2 = new5.sum()

new_df2 = purchase_count2.copy()
for i in list(purchase_count2.index.values):
    
    new_df2.loc[i,"Average Purchase Price"] = average_purchase_p2.loc[i, "Price"]
    new_df2.loc[i,"Total Purchase Value"] = total_purchase_v2.loc[i, "Price"]
    new_df2.loc[i,"Avg Total Purchase per Person"] = total_purchase_v2.loc[i, "Price"]/unique_name_count.loc[i, "Total Count"]
    
# new_df2.dtypes
```

```python
new_df2["Average Purchase Price"] = new_df2["Average Purchase Price"].map("${:.2f}".format)
new_df2["Total Purchase Value"] = new_df2["Total Purchase Value"].map("${:.2f}".format)
new_df2["Avg Total Purchase per Person"] = new_df2["Avg Total Purchase per Person"].map("${:.2f}".format)
new_df2
```

```python
# Q6

# "filted" the date to keep targeted values
new6 = data_file.copy()
new6 = new6.groupby("SN")


# calculate the values
purchase_count3 = new6["SN"].count()
total_purchase_value3 = new6["Price"].sum()
average_purchase_p3 = new6["Price"].mean()

# merge the data
new_df2 = pd.DataFrame(total_purchase_value3)
new_df2["Purchase Count"] = purchase_count3
new_df2["Average Purchase Price"] = average_purchase_p3

new_df2 = new_df2.rename(columns={"Price":"Total Purchases"})

new_df2.sort_values(by="Total Purchases", ascending=False, inplace=True)
new_df2["Total Purchases"] = new_df2["Total Purchases"].map("${:.2f}".format)
new_df2["Average Purchase Price"] = new_df2["Average Purchase Price"].map("${:.2f}".format)
new_df2.head(5)
```

```python
# Q7

new7 = data_file.copy()
new7 = new7.groupby(["Item ID", "Item Name"])

purchase_count3 = new7["Item ID"].count()
item_price = new7["Price"].unique()
item_price = item_price.astype(float)
total_purchase_v4 = new7["Price"].sum()

new_df4 = pd.DataFrame(purchase_count3)
new_df4 = new_df3.rename(columns={"Item ID":"Purchase Count"})

new_df4["Item Price"] = item_price
new_df4["Total Purchase Value"] = total_purchase_v4
# new_df4

new_df5 = new_df4.sort_values(by="Purchase Count", ascending=False)
new_df5["Total Purchase Value"] = new_df4["Total Purchase Value"].map("${:.2f}".format)
new_df5["Item Price"] = new_df4["Item Price"].map("${:.2f}".format)

new_df5.head(5)
```
